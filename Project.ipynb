```python
display(data_covid2.head(2),data_covid2.tail(2))
```

ข้อมูลชุดที่ 3 รายงานจำนวนผู้ติดเชื้อ COVID-19 ประจำวัน ตั้งแต่วันที่ 25ตุลาคม 2564 ถึงวันที่ 27 กุมภาพันธ์ 2565

```python
data_file_path = os.path.join(path,'confirmed-cases-since-271064.csv')   # ดึงไฟล์จากไดร์ฟที่เราต้องการ
data_covid3 = pd.read_csv(data_file_path)   # อ่านไฟล์ csv
```

```python
display(data_covid3.head(2),data_covid3.tail(2))
```

ข้อมูลชุดที่ 4 รายงานจำนวนผู้ติดเชื้อ COVID-19 ประจำวัน ตั้งแต่วันที่ 28 กุมภาพันธ์ 2565 ถึงวันที่ 20 มีนาคม 2565

```python
data_file_path = os.path.join(path,'confirmed-cases-since-280265.csv')   # ดึงไฟล์จากไดร์ฟที่เราต้องการ
data_covid4 = pd.read_csv(data_file_path)   # อ่านไฟล์ csv 
```

```python
display(data_covid4.head(2),data_covid4.tail(2))
```

ขั้นตอนถัดไป คือ การรวมข้อมูลารางทั้ง 4 ชุด เข้าด้วยกัน และจัดเรียงตามลำดับก่อน-หลัง

```python
data_covid_all =pd.concat([data_covid1,data_covid2,data_covid3,data_covid4], ignore_index=True)
data_covid_all
```

เช็คดู ค่า Missing

```python
data_covid_all.isnull() #เช็คดูค่า Missing ทั้งหมด ใน Data frame
```

```python
data_covid_all.isnull() .any() # เช็คดูค่า Missing ในแต่ละ คอลัมป์ 
```

ขั้นตอนต่อมาคือการทำความสะอาดข้อมูล Data cleanning

**1.เลือกเฉพาะฟิวด์ที่ใช้ ได้แก่ announce_date,	sex,	age	,Unit,	nationality risk	และ province_of_onset**

```python
data_p1 = data_covid_all[['No.','announce_date','sex','age','Unit','nationality','province_of_onset','risk']]
data_p1
```

**2.ฟิวด์ age เปลี่ยนค่า Nan ให้เป็นค่าเฉลี่ย ของอายุ**

```python
data_covid2[['age']].mean() # หาค่าเฉลี่ยของอายุ พบว่าอายุเฉลี่ย ประมาณ 35 ปี 
```

```python
data_covid2[['age']].describe()
```

```python
 data_p1[['age']] = data_p1[['age']].fillna('35.0') #เปลี่ยนค่า Nan ให้เป็นค่าเฉลี่ย ของอายุ นั่นคือ 35.0
```

```python
data_p1.isnull() .any() # พบว่า ฟิวด์ของ age เป็นค่า false แล้ว 
```

```python
data_p1
```

**3.เปลี่ยน ข้อมูลในฟิวด์  Unit ที่มีค่าเป็น NaN ให้เป็น คำว่า "ปี"**

```python
 data_p1[['Unit']] = data_p1[['Unit']].fillna('ปี')
```

```python
data_p1.isnull().any() 
```

```python
data_p1.info()
```

4.เปลี่ยน ข้อมูลในฟิวด์ sex  nationality province_of_onset	และrisk ที่มีค่าเป็น NaN ให้เป็น คำว่า "ไม่ระบุ"

```python
 data_p1[['sex']] = data_p1[['sex']].fillna('ไม่ระบุ')
```

```python
 data_p1[['nationality']] = data_p1[['nationality']].fillna('ไม่ระบุ')
```

```python
data_p1[['province_of_onset']] = data_p1[['province_of_onset']].fillna('ไม่ระบุ')
```

```python
 data_p1[['risk']] = data_p1[['risk']].fillna('ไม่ระบุ')
```

```python
data_p1 
```

```python
data_p1.isnull().any() 
```

✔️**พบว่าไม่มีค่า Missing แล้ว แสดงว่าข้อมูลมีความสะอาดมากขึ้น พอที่จะนำไปวิเคราะห์ต่อในขั้นตอนถัดไปออกแบบและสร้างตาราง pandas dataframe ที่ สามารถนำมาดึงข้อมูลด้วย pandas ได้  ****
