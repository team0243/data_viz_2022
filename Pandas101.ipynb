HW (ให้ใช้เฉพาะที่อาจารย์สอนไปแล้วในวิชานี้)

1.   คำนวณ อายุเฉลี่ย ของผู้ชาย ของข้อมูลทั้งหมด
2.   คำนวณ อายุเฉลี่ย ของผู้ชาย ของผู้ป่วยในจังหวัดขอนแก่น

3. หาจำนวนผู้ป่วยที่เป็นคน "**คนต่างชาติเดินทางมาจากต่างประเทศ**"

````
import math 
````

````
this_data = data_covid[['sex','age','province_of_onset']] #เลือกนำเข้าข้อมูลที่ต้องใช้ คือ เพศ อายุ province_of_onset
````

````
this_data
````

````
m = this_data[this_data['sex']=='ชาย']  #เลือกข้อมูลของผู้ชาย นำมาคำนวณ
m
````

### ข้อ 1

````
sum = 0 #ถ้าไม่มีอะไรมาบวกให้เป็น 0 ก็คือเริ่มต้นที่ 0 ตามสมบัติการบวกเลขคณิต
N = 0 # 
for a in m['age']: #วนลูปอายุ
    
    if a > 0:  #ถ้าอายุมากกว่า 0 
        sum += a # sum = sum+a (นำมารวมกัน)
        N += 1

print(f'อายุเฉลี่ย ของ ผู้ป่วยชายเท่ากับ{sum/N} ปี หรือคิดเป็นอายุประมาณ {math.floor(sum/N)} ปี')
````

### ข้อ 2

````
mkk= m[m['province_of_onset']=='ขอนแก่น'] # สนใจข้อมูลที่มีเฉพาะจังหวัดขอนแก่น ที่มีผู้ป่วยชาย 

sum = 0   # ค่าทั้งหมดมาบวกกันไปเรื่อย ๆ # ดูจากค่า identity ของการบวกเเละการคูณ
N = 0
for a in mkk['age']:
    
    if a > 0:
        sum += a  # sum = sum + a 
        N += 1   # บวก 1 ขึ้นไปเรื่อย ๆ 

print(f'อายุเฉลี่ย ของ ผู้ป่วยชายในจังหวัดขอนแก่นเท่ากับ {sum/N} ปี หรือคิดเป็นอายุประมาณ {math.floor(sum/N)} ปี')
````

*  พบว่าอายุเฉลี่ยของ ผู้ป่วยชายในจังหวัดขอนแก่นคิดเป็นอายุเท่ากับ 26.571428571428573 ปี ซึ่งมีอายุเฉลี่ยน้อยกว่า อายุเฉลี่ย ของ ผู้ป่วยชายทั่วประเทศ เท่ากับ39.46076940331987ปี

### การวนลูป record ในตาราง  (```.iterrows```)

````
this_data = data_covid[['sex','age','province_of_onset']]
````

````
for each_row in this_data.iterrows():
    # print(each_row)
    # print(each_row[1])
    if (each_row[1]['age'] == 20) and (each_row[1]['province_of_onset'] == 'ขอนแก่น'):
        print(each_row)
````

### การวนลูป แบบมองตารางแพนด้าส์(pandas dataframe) เป็น numpy array  (```.iloc```)

````
for each_row in range(this_data.shape[0]):
    if (this_data.iloc[each_row,1] == 20) and (this_data.iloc[each_row,2] == 'ขอนแก่น'):  #each_row คือ ตัวเลข
        print(each_row)
        print(this_data.iloc[each_row,:])   
````
