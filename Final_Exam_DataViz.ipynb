# alldb

```python
alldb = pd.read_csv('alldb.csv')
print(alldb.shape)
alldb.head()
```

```python
alldb = alldb.drop(columns=['Unnamed: 0','care','pid'])     # ลบ คอลัมน์ที่ไม่ต้องการออก
print(alldb.shape)
alldb.head()
```

```python
alldb.isnull().sum()
```

## C: user

user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['user'].astype(str).str.isdigit()) & (alldb['user'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ user ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
print(alldb.shape)                                                                             # .str.isdigit ใช้ตรวจสอบว่าอักขระทั้งหมดเป็นตัวเลขหรือไม่
alldb.head()
```

ref2: https://stackoverflow.com/questions/24883503/pandas-select-only-numeric-or-integer-field-from-dataframe

```python
alldb.isnull().sum()      # ดูค่าว่าง
```

## C: date

date เป็น YYYY-MM-DD



```python
alldb['date'] = pd.to_datetime(alldb['date'],format="%Y-%m-%d")     # เปลี่ยนชนิดข้อมูล และจัดรูปแบบข้อมูล
print(alldb.shape)
alldb.head(3)
```

```python
alldb['date'] = alldb[['date']].fillna(method="ffill")      # เติมค่าว่างของวันที่ด้วย ค่าก่อนหน้า หรือค่าข้างบน
print(alldb.shape)
alldb.head(3)
```

```python
print(alldb.dtypes)
print('*'*50)
alldb.isnull().sum()
```

## C: time

time เป็น hh-mm-ss

```python
alldb['time'] = alldb[['time']].fillna(method="ffill")      # เติมค่าว่างของเวลาด้วย ค่าก่อนหน้า หรือค่าข้างบน
print(alldb.shape)
alldb.head(3)
```

```python
alldb.isnull().sum()
```

## C: age

```python
alldb['age'] = alldb['age'].astype(float).astype('Int64')       # แปลงข้อมูล จาก object เป็น int ด้วยการ แปลงจาก object -> float -> int64
```

```python
print(alldb.dtypes)
```

```python
outlier = alldb[(alldb['age'] < 1) | (alldb['age'] > 150)] # เลือกอายุที่น้อยกว่า 1 หรือ มากว่า 150 ปี
print(outlier.shape)
outlier.head()
```

```python
alldb[['age']] = alldb[['age']].replace(dict.fromkeys(set(outlier['age']), pd.NA))      # เปลี่ยนค่าผิดปกติเป็นค่าว่าง
print(alldb.shape)
alldb.head(3)
```

```python
alldb.isnull().sum()
```

```python
mean_age = alldb['age'].mean()      # หาอายุเฉลี่ย
mean_age
```

```python
alldb['age'] = alldb['age'].fillna(54)     # แทนค่า nan ของ age ด้วย mean
```

```python
print(alldb.dtypes)
print('*'*50)
alldb.isnull().sum()
```

```python
print(f'เสียข้อมูลไป {40035-38171} แถว')
print(f'คิดเป็น {((40035-38171)*100)/40035} %')
```
